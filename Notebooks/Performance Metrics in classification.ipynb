### Import libraries.

```python
import pandas as pd 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
```

```python
fraud_df = pd.read_csv('../Datasets/creditcard.csv')
```

### Dataset overview.

```python
fraud_df.sample(5)
```

```python
fraud_df.info()
```

```python
# split the dataset.
```

```python
X = fraud_df.iloc[ : , : 30]
y = fraud_df['Class']
```

```python
X.sample()
```

```python
y
```

```python
# train test split. 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
print(X_train.shape) 
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# train the models.
```

```python
# train logistic regression.

lr_clf = LogisticRegression(max_iter=500)
lr_clf.fit(X_train, y_train)
```

```python
lr_y_pred = lr_clf.predict(X_test)
```

```python
# train the Decision treee model. 

dt_clf = DecisionTreeClassifier(max_depth=100, max_features=30)
dt_clf.fit(X_train, y_train)
```

```python
dt_y_pred = dt_clf.predict(X_test)
```

```python
# accuracy score of both the models. 
print('Accurcay score of LR : ', accuracy_score(y_test, lr_y_pred))
print('Accurcay score of DT : ', accuracy_score(y_test, dt_y_pred))
```

```python
# Precsion of both the models. 
print('Precsion score of LR : ', precision_score(y_test, lr_y_pred))
print('Precsion score of DT : ', precision_score(y_test, dt_y_pred))
```

```python
# Recall of both the models. 
print('Recall score of LR : ', recall_score(y_test, lr_y_pred))
print('Recall score of DT : ', recall_score(y_test, dt_y_pred))
```

```python
# Consfusion matrix of both the models. 
```

```python
print('Confusion matrix score of LR : ')
print(confusion_matrix(y_test, lr_y_pred))
```

```python
print('Confusion matrix score of DT : ')
print(confusion_matrix(y_test, dt_y_pred))
```

```python
# Precsion of both the models. 
print('Precsion score of LR : ', precision_score(y_test, lr_y_pred, average=None))
print('Precsion score of DT : ', precision_score(y_test, dt_y_pred, average = None))
```

`NOTE : ` It will always take the value towards the lower value in precsion and recall as show in the above cell.

`Multi-class classification : ` In multi-class classification we can also calculate the weighted and Macro values of all precsion, recall and f1-score.
